```python
###########################
#####  RAKE Method  #######
###########################
#### Find keywords and phrases
'''
Rapid Automatic Keyword Extraction is an algorithm to extract keywords and phrases.
It works on the assumptions that 1) keywords often contain multiple words, 2) are separated by
punctuation and stopwords.

Using punctuation characters and stopwords as parsers, a body of text is partitioned into possible
keywords.

Scores for these possible keywords are calculated using word co-occurance
(def: degree and count of how two or more words occur together in a text body)

(Deg (w)) favors words that occur often and in longer candidates.
(Freq (w)) favors words that occur frequently regardless of the number of words with which they co-occurred.
(Deg (w)/ Freq (w) favors the words that predominately occur in longer candidate keywords.

Evalute the keyword extraction using precision, recall, and F score
'''

#process:
#1. import RAKE
#2. import stopwords
#3. run RAKE on text
#4. sort by score DESC
#5. take top X keywords/phrases
import RAKE
import operator
```
